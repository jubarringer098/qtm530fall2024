(j) Plot by category

<font size = "3">

- Read "data_raw/wdi_urban.csv"<br>
This contains data from 1980 and 2020

- Create a list with the unique values of "year"
- Create a plot with a ...
    - Histogram of "prop_urbanpopulation" <br>
    for each year
    - Set the histogram option ``` alpha = 0.5 ``` for transparency
    - All the histograms should be overlapping
    - Create a legend with the list of years
    - Label the axes and the title

What do we learn from the graph?

HINT: See the example at the end of Lecture 10


````python
# Write your own code

wdi_urban = pd.read_csv("data_raw/wdi_urban.csv")
# display(wdi_urban)

list_unique_years = pd.unique(wdi_urban["year"])
# print(list_unique_years)

for category in list_unique_years:
    df = wdi_urban.query("year == @category")
    plt.hist(x = df["prop_urbanpopulation"], alpha = 0.5)

plt.xlabel("prop_urbanpopulation")
plt.ylabel("Frequency")
plt.title("Histogram of Urban Population Proportion by Year")
plt.legend(labels = list_unique_years, title  = "Years")

plt.show()


# We learn that there's been an overall increase in urban population proportion from 1980 to 2020; the urban population proportions seem to be more concentrated in higher ranges in 2020 than in 1980. 
````

(k) Composite operations on arrays

<font size = "3">

- Define $r = 0.04$
- Define two arrays $A$ and $T$ using "np.array()", assigning the following values

$ a = \begin{pmatrix} a_1 \\ a_2 \\ a_3 \end{pmatrix} =  \begin{pmatrix} 20 \\ 30 \\ 15 \end{pmatrix},
\qquad  t = \begin{pmatrix} t_1 \\ t_2 \\ t_3 \end{pmatrix} = \begin{pmatrix} 0 \\ 1 \\ 2 \end{pmatrix} $

- Use element-by-element vector operations to compute a new array $M$ of length 3:

$ M = \begin{pmatrix} a_1 / (1+r)^{t_0}  \\ a_2 / (1+r)^{t_1} \\ a_3 / (1+r)^{t_2} \end{pmatrix} $

- Compute the sum of the values of $M$ using "M.sum()"

Note: I am using subscripts to denote the individual elements of $a$ and $t$.
- In the code just enter the values of $a$ and $t$.
- To get full points you must do all of the above tasks



````python
# Write your own code

r = 0.04

A = np.array([20, 30, 15])
t = np.array([0, 1, 2])

# I'm assuming that the t's should be t1, t2, t3 and not t0, t1, and t2

M = A/(1+r)**t

sum_M = M.sum()

print(sum_M)





````
